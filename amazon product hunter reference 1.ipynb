```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager

from selenium import webdriver
import selenium
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options

from os import listdir
from os.path import isfile, join
import pandas as pd
import time
import os
import requests
import lxml.html as lh
try: 
    from BeautifulSoup import BeautifulSoup as bs
except ImportError:
    from bs4 import BeautifulSoup as bs
```

```python
url_input=input("Enter amazon search url")
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
driver.get(url_input)
try:
    driver.maximize_window()
except:
    pass
```

```python
page=[]
url=[]
item={'Title':[],'image':[],'MRP':[]}

try:
    next_page = driver.find_element_by_partial_link_text('Next')
except:
    next_page=True

time.sleep(5)
page_last=7
i=0
brk=False
while not next_page == False:
#     if brk==True:
#         break
    if i> page_last:
        break
    i=i+1
    for cell in driver.find_elements_by_class_name('s-result-item'):
        image_array=[]
        try:
            htmlText = cell.get_attribute('innerHTML')
            image_index=htmlText.find('srcset="')
            image_array=htmlText[image_index:].split('"')[1].replace(",",' ').split(' ')
            try:
                for i in range(len(image_array)):
                    if len(image_array[i])<5 or image_array==None:
                        del image_array[i]
                    elif  not image_array[i][:4] ==  'http':
                        del image_array[i]
            except:
                pass
        except:
            pass
        if len(image_array)>0:
            print(cell.text)
            txt=cell.text.split('\n')
            if txt[0]=='Sponsored':
                del txt[0]
            
            try:
                price_index=txt[2].find("₹")
                sale=txt[2][price_index:].split(' ')[0]
                sale = int(sale.replace('₹',''))
            except:
                continue
            try:
                price_index=txt[3].find("₹")
                mrp=txt[3][price_index:].split(' ')[0]
                if mrp == 'A':
                    mrp=sale
                mrp = int(mrp.replace('₹',''))
            except:
                mrp=sale
            if mrp>sale:
                mrp=sale
            item['Title'].append(txt[0])
            item['MRP'].append(mrp)
            item['image'].append(image_array)
            print('images:',image_array)
#     if brk==True:
#         break
    try:
        next_page = driver.find_element_by_partial_link_text('Next').click()
    except:
        next_page=False
    page.append(driver.page_source)
    url.append(driver.current_url)
    time.sleep(5)
```
