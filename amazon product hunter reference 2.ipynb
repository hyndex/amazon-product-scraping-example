```python
url=input('Please enter a url')
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get(url)
driver.maximize_window()
list_url=[]
try:
    last_page = driver.find_element_by_css_selector('#search > div.s-desktop-width-max.s-desktop-content.sg-row > div.sg-col-20-of-24.sg-col-28-of-32.sg-col-16-of-20.sg-col.sg-col-32-of-36.sg-col-8-of-12.sg-col-12-of-16.sg-col-24-of-28 > div > span:nth-child(4) > div.s-main-slot.s-result-list.s-search-results.sg-row > div:nth-child(31) > span > div > div > ul > li.a-disabled')
except:
    last_page = driver.find_element_by_css_selector('#pagn > span.pagnDisabled')
last_page = int(last_page.text)
for i in range(last_page-1):
    try:
        next_page = driver.find_element_by_css_selector('#pagnNextString')
    except:
        next_page = driver.find_element_by_partial_link_text('Next')
    list_url.append(driver.current_url)
    next_page.click()
    time.sleep(5)
```
